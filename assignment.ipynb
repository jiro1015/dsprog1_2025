# 最終課題

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse
import time

base_url = "https://www.musashino-u.ac.jp/"
visited = set()
sitemap = {}

MAX_DEPTH = 2         
MAX_PAGES = 200

def crawl(url, depth=0):
    if url in visited or depth > MAX_DEPTH or len(sitemap) >= MAX_PAGES:
        return
    visited.add(url)
    
    try:
        time.sleep(1)
        response = requests.get(url, timeout=10)
        response.raise_for_status()
    except Exception as e:
        print(f"アクセス失敗: {url} ({e})")
        return
    
    soup = BeautifulSoup(response.text, "html.parser")
    title_tag = soup.find("title")
    title = title_tag.get_text(strip=True) if title_tag else "（タイトルなし）"
    sitemap[url] = title
    print(f"[{len(sitemap)}] {url} - {title}")
    
    for link in soup.find_all("a", href=True):
        new_url = urljoin(url, link["href"])
        if urlparse(new_url).netloc.endswith("musashino-u.ac.jp"):
            crawl(new_url, depth + 1)

crawl(base_url)
for url, title in sitemap.items():
    print(f"{url} : {title}")

print(f"\n合計 {len(sitemap)} ページを取得しました。")
```
